```python
import json
import matplotlib.pyplot as plt

# Load JSON data from the provided sample file
with open('./outputs_FPN_VisDrone/metrics.json', 'r') as f:
    data = json.load(f)

# Print keys of the first entry to see what keys are available
print("Available keys:", list(data[0].keys()))

# Extracting data for plotting
iterations = [entry['iteration'] for entry in data]

# Check if the key exists before extracting its value
loss_cls = [entry['loss_cls'] if 'loss_cls' in entry else None for entry in data]
loss_box_reg = [entry['loss_box_reg'] if 'loss_box_reg' in entry else None for entry in data]
total_loss = [entry['total_loss'] if 'total_loss' in entry else None for entry in data]

# Plotting the data
plt.figure(figsize=(10, 6))
plt.style.use('classic') 
plt.plot(iterations, loss_cls, label='Loss_cls', marker='o')
plt.plot(iterations, loss_box_reg, label='Loss_box_reg', marker='x')
plt.plot(iterations, total_loss, label='Total_loss', marker='s')

plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Losses Across Iterations')
plt.legend()
plt.grid(True)
plt.show()
```

### <span style="font-weight:normal;">Visualizing the progression of different loss metrics over iterations during the training process. The x-axis represents the iteration number, indicating the progress of the training. The y-axis represents the loss values, which are indicative of the deviation between predicted and ground truth values.</span>

```python
from collections import defaultdict

# Load JSON data from the provided file
with open('./outputs_FPN_VisDrone/inference/coco_instances_results.json', 'r') as f:
    data = json.load(f)

# Aggregate scores by category_id
scores_by_category = defaultdict(list)
for entry in data:
    category_id = entry['category_id']
    score = entry['score']
    scores_by_category[category_id].append(score)

# Calculate the mean score for each category_id
mean_scores = {category_id: sum(scores) / len(scores) for category_id, scores in scores_by_category.items()}

# Sort the mean scores by category_id for plotting
sorted_mean_scores = dict(sorted(mean_scores.items()))

# Plot the aggregated scores as a bar plot
plt.figure(figsize=(12, 6))
plt.bar(sorted_mean_scores.keys(), sorted_mean_scores.values(), color=plt.cm.tab20b.colors)
plt.xlabel('Category ID')
plt.ylabel('Mean Score')
plt.title('Mean Scores by Category ID')
plt.tight_layout()
plt.show()
```

### <span style="font-weight:normal;">Visualizing the average scores obtained for each category ID in the dataset. The x-axis represents the category IDs, while the y-axis represents the mean scores. Each bar corresponds to a specific category ID, with the height of the bar indicating the average score obtained for that category.</span>
